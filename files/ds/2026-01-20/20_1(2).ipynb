```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv('train_14.csv',usecols=['GarageQual','FireplaceQu','SalePrice'])
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde',ax=ax)
df[df['GarageQual'].isnull()]['SalePrice'].plot(kind='kde',ax=ax,color='r')

lines,labels=ax.get_legend_handles_labels()
labels=['Houses with TA','Houses with NA']
ax.legend(lines,labels,loc='best')

plt.title('GarageQual')
```

```python
df['GarageQual'].isnull().sum()
```

```python
df['GQ']=df['GarageQual'].fillna('TA')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde',ax=ax)
df[df['GQ']=='TA']['SalePrice'].plot(kind='kde',ax=ax,color='r')

lines,labels=ax.get_legend_handles_labels()
labels=['OG value','Imputed value']
ax.legend(lines,labels,loc='best')

plt.title('GarageQual')
```
