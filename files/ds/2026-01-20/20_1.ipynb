```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_of_dataset = "titanic_1.csv"
df= pd.read_csv(path_of_dataset)
```

```python
df.info
```

```python
df.isnull().mean()*100
```

```python
x=df.drop(columns=['Survived'])
y=df['Survived']
```

```python
mean_age=x['Age'].mean()
```

```python
median_age=x['Age'].median()
```

```python
x['Age_mean']=x['Age'].fillna(mean_age)
x['Age_Median']=x['Age'].fillna(median_age)
```

```python
x
```

```python
x.isnull().sum()
```

```python
x['Fare'].isnull().sum()
```

```python
mean_fare=x['Fare'].mean()
median_fare=x['Fare'].median()
```

```python
x['Fare_mean']=x['Fare'].fillna(mean_fare)
x['Fare_Median']=x['Fare'].fillna(median_fare)
```

```python
x
```

```python
x.isnull().sum()
```

```python
fig=plt.figure()
ax = fig.add_subplot(111)

x['Age'].hist(bins=50,ax=ax,density=True,color='red')

x['Age_mean'].hist(bins=50,ax=ax,color='green',density=True,alpha=0.8)
```

```python
print("Og age variable variance:",x['Age'].var())
print("age variance after median imputation:",x['Age_mean'].var())
print("age variance after mean imputation:",x['Age_Median'].var())
print("Og fare variable variance:",x['Fare'].var())
print("fare variance after median imputation:",x['Fare_mean'].var())
print("fare variance after mean imputation:",x['Fare_Median'].var())
```
