# Need for plotting

*Data visualization is a part of the data preprocessing step as well as the data presentation step.*

What is the need to visualize data when there are numerical values already present?
- Easier to compare plotted values than numeric values.
- Gain better intuition of the data than would be possible by looking at tables of values alone.
- Hidden patterns may get highlighted which may be used later for model selection.

How do we visualize data in Python?
- Mainly done using libraries such as matplotlib and seaborn where seaborn is built on top of matplotlib.
- Using seaborn library, a lot can be achieved with the help of fewer lines of code.

## About the dataset

- Anscombe dataset (also known as Anscombe's quartet).
- Created by English statistician Frank Anscombe.
- Purpose is to show the importance of statistical graphs, thereby showing the importance of visualization.
- Contains four sets of data, each of which contains two continuous variables.
- Each set has the same values of statistical measures.
- However, data visualization reveals that each set does not follow the same pattern.
- Pitfalls of looking at only summary statistics are also highlighted.
- Such is the popularity of the dataset that it is built into the seaborn library.

## Basic subplots

*Instead of creating individual plots, it is better to create subplots in a single figure. Less code, more compact.*

```python
# defining the plotting area is the first step
figure1 = plt.figure()

# need 4 subplots inside the main figure
# numbered left-to-right, top-to-bottom starting from 1

# subplot location 1 (figure 1, subplot 1; so two 1s in variable name)
axes11 = figure1.add_subplot(2, 2, 1)

# subplot location 2
axes12 = figure1.add_subplot(2, 2, 2)

# subplot location 3
axes13 = figure1.add_subplot(2, 2, 3)

# subplot location 4
axes14 = figure1.add_subplot(2, 2, 4)

# Use the plot method on each axis to create subplots and as a result the entire plot
axes11.plot(df_anscombe_1['x'], df_anscombe_1['y'], 'o')
axes12.plot(df_anscombe_2['x'], df_anscombe_2['y'], 'o')
axes13.plot(df_anscombe_3['x'], df_anscombe_3['y'], 'o')
axes14.plot(df_anscombe_4['x'], df_anscombe_4['y'], 'o')

# Add title to each subplot
axes11.set_title("First dataset")
axes12.set_title("Second dataset")
axes13.set_title("Third dataset")
axes14.set_title("Fourth dataset")

# Add X-axis label to each subplot
axes11.set_xlabel("X")
axes12.set_xlabel("X")
axes13.set_xlabel("X")
axes14.set_xlabel("X")

# Add Y-axis label to each subplot
axes11.set_ylabel("Y")
axes12.set_ylabel("Y")
axes13.set_ylabel("Y")
axes14.set_ylabel("Y")

# Add title to the entire figure (plot)
figure1.suptitle("Anscombe data")

# The axes should be spread apart from one another
figure1.tight_layout()
```

## Univariate plots

*In Statistics, the term "univariate" refers to a single variable.*

### Histogram

*This is the most common means of looking at a single variable.*

```python
# defining the plotting area is the first step
fig2 = plt.figure()

# adding a subplot to the created figure
axes21 = fig2.add_subplot(1, 1, 1)

# plotting a histogram in the subplot of the figure
bin_density, bin_edge_values, patches = axes21.hist(df_tips['total_bill'], bins=10)

# showing all the intervals (bins)
plt.xticks(bin_edge_values, rotation=45)

# setting the title of the histogram
axes21.set_title('Histogram of total bill')

# setting the X-axis label; the variable whose frequency is to be found
axes21.set_xlabel("Total bill")

# setting the Y-axis label; the frequency itself
axes21.set_ylabel("Frequency")

# displaying the figure containing the subplot with the histogram
fig2.show()

# saving the figure as a vector graphic with high resolution for later usage
fig2.savefig("Histogram of total bill.svg", format='svg', dpi=300)
```

*Why extract bin_density and bin_edge_values?*

```python
# number of values in each group / interval / bin
# OR
# frequency of each bin
print(bin_density)
```

**Interpretation**: *The histogram shows a right-skewed distribution of total bills, with most bills concentrated in the lower to mid range (around $1020) and relatively few high-value bills.*

```python
# same histogram plot using seaborn
# creating a figure and adding a subplot to the figure using a single line of code
# control entire figure details using fig3
# control subplot details using ax31

fig3, axes31 = plt.subplots()

axes31 = sns.histplot(df_tips['total_bill'])  # not specifying bins?
axes31.set_title("Histogram of total bill")
axes31.set_xlabel("Total bill")
axes31.set_ylabel("Frequency")

plt.show()
```

### Density plot (KDE: Kernel Density Estimation)

*Used for univariate plotting where normal distributions are drawn centered by each data point and then smoothing out the overlapping plots so that the area under the curve is 1.*

```python
fig4, axes41 = plt.subplots()

axes41 = sns.kdeplot(df_tips['total_bill'])
axes41.set_xlabel("Total bill")
axes41.set_ylabel("Density")
axes41.set_title("KDE plot of total bill")

plt.show()
```

**Interpretation**: *Most bills cluster around the range (highest density) 15-20. This is the most common spending range. The distribution is right-skewed, with a long tail toward higher bills, meaning large bills are less frequent but do occur. Very small bills are rare, and the smooth curve suggests no strong multimodalityspending is concentrated around a single typical range.*

### Scatter plot

*Scatter plots are used when a continuous variable is plotted against another continuous variable.*

### Hexbin plot

**Drawback of scatter plot**: Not meaningful if there are too many points.

*Better approach is to "bin" points on the plot together. Just as histograms can bin a variable to create a bar, hexbin plot can bin two variables using a hexagon.*

```python
joint_hexbin = sns.jointplot(x="total_bill", y="tip", data=df_tips, kind="hex")
joint_hexbin.set_axis_labels(xlabel="Total bill paid", ylabel="Tip paid")
joint_hexbin.fig.suptitle("Hexbin joint plot of total bill and tip", fontsize=10, y=1.03)

plt.show()
```

### 2D KDE plot

*A 2-D KDE plot is a smooth heatmap of point density showing where combinations of two variables occur most often.*

Use KDE plots for/when:
- Large datasets where scatter plots overlap too much.
- Focus is on density patterns and not individual points.

```python
fig81, axes811 = plt.subplots()

axes811 = sns.kdeplot(x="total_bill", y="tip", data=df_tips, fill=True, ax=axes811)

axes811.set_xlabel("Total bill paid")
axes811.set_ylabel("Tip offered")
axes811.set_title("2D KDE plot: Total bill vs Tip")

plt.show()
```

**Interpretation**: *The 2D KDE plot shows a strong concentration of observations where total bills are around 1020 and tips are around 23, indicating this is the most common billingtipping combination. The density contours slope upward, showing a positive relationship: as the total bill increases, the tip generally increases. Higher bills with larger tips exist but are less frequent, as indicated by the lighter outer contours. Overall, tipping behavior scales with bill size, with most customers clustered in a moderate spendingtipping range.*

### Box plot

*Box plots are used when a continuous variable is plotted against a discrete variable.*

```python
# defining the plotting area is the first step
figure4 = plt.figure()

axes41 = figure4.add_subplot(1, 1, 1)

# data has to be passed as a list; tip paid by female customers and tip paid by male customers
axes41.boxplot([df_tips[df_tips['sex']=='Female']['tip'], df_tips[df_tips['sex']=='Male']['tip']],
               tick_labels = ['Female', 'Male'])

axes41.set_xlabel('Gender of the tip provider')
axes41.set_ylabel('Tip')
axes41.set_title('Box plot of tips paid with respect to gender')

figure4.show()

figure4.savefig("boxplot.svg", format="svg", dpi=300)
```

**Interpretation**:
- The box plot shows the central tendency, spread, and presence of outliers in the data.
- The orange line denotes the second quartile (Q2) or 50th percentile of the data; also known as median.
- The lower edge of the box denotes first quartile (Q1) or 25th percentile of the data.
- The upper edge of the box denotes third quartile (Q3) or 75th percentile of the data.
- The interquartile range (IQR) is given by the difference of Q3 and Q1.
- The whisker towards the lower edge denotes the smallest value which is not an outlier.
- The whisker towards the upper edge denotes the highest value which is not an outlier.
- The points beyond the whisker towards the upper edge highlight the presence of outliers, suggesting extreme values that deviate significantly from the majority of the data.

- The median tip (shown by the white dash) is slightly higher for males than for females, indicating marginally higher tips by males.
- The spread of tips is larger for male customers, with a longer upper tail, suggesting that very high tips occur more often among males.
- Female customers show a more compact distribution, meaning their tip amounts are more consistent with fewer extreme values.

**Summary**: *While typical tip amounts are similar between genders, male customers exhibit greater variability and a higher likelihood of unusually large tips.*
